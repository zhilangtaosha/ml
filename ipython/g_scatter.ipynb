```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# データ準備

```python
tips = sns.load_dataset("tips")
tips[tips.total_bill== 16.99] #ただ使っただけ
tips.head(3)
```

# 散布図

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.scatter, "total_bill", "tip", edgecolor="w")
```

```python
g = sns.FacetGrid(tips, col="time",  hue="smoker")
g = (g.map(plt.scatter, "total_bill", "tip", edgecolor="w").add_legend())
```

# 散布図の意味

- 2つの要素からなる1組のデータが得られたときに、2つの要素の関係を見るためにプロットしたグラフ
- 特性と要因との関係を調べたいときに使用
- 異常点がある場合は、原因を調べ、その原因が明確であれば除く。原因が不明の場合は、その点を含めて判断。
- 全体として相関がない場合は、層別でも確認。
