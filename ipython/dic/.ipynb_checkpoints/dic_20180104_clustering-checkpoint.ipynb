# 卸売業者 教師なし学習

**【課題】教師なし学習とは**
- 教師あり学習との比較

    教師なし学習とは、正解となる情報がない状態でモデルを作り上げることを指す。
    教師あり学習は、その反対に正解となる情報である教師データに基づいて、計算を繰り返し、重み付けをする係数を更新しながらモデルとなる関数や分布を作り上げる子を指す。


- どんなときに使うのか

    教師なし学習は、画像のようなデータから特徴をつかみ、似たような要素から構成するグループに分けるようなアルゴリズムとして使われる。また、データを構成する部品、もしくは成分を見つけるアルゴリズムとしても扱われる。


- 具体的にどんなものがあるのか

 - 似たようなデータからグループを構成するアルゴリズム
   - 画像のセグメンテーションを使った医療画像処理 \*1
   - 物体認識
 - データから成分を見つけるアルゴリズム
   - 系列やパターンのマイニングといったデータマイニングの分野
     - マーケティング\*2。年齢や性別、居住地など、よく似た属性の組み合わせを持つカスタマーをセグメント化して、教師なし学習のアルゴリズムで購入情報や閲覧情報を学習させることで、カスタマーの嗜好性を分析したリコメンデーション機能 
   - シーケンス解析や遺伝子クラスタリングといったバイオインフォマティックス、
   - 系列やパターンのマイニングといったデータマイニングの分野


\*1 [教師なし学習](https://jp.mathworks.com/discovery/unsupervised-learning.html)

\*2 [そういうことだったのか！と思わずうなずく「機械学習」超入門](https://ferret-plus.com/6215)

# Feature Scaling

**【課題】Feature Scalingの実行**

```python
#dataに対して適切なFeature Scalingを行うコードを記述

data_fs=np.log1p(data)
```

```python
pd.scatter_matrix(data_fs, alpha = 0.3, figsize = (14,8), diagonal = 'kde')
```

**【課題】なぜFeature Scalingが必要なのか記述せよ**

- 同じものさしで数値を見ないと真実が見えない場合がある。[違う指標のものを比べたりする場合に単位が違う効果を打ち消す必要がある。](http://retrofocus.hatenablog.com/entry/2015/05/02/151526) 例えばテストの合計点。
  - Eng: 10,60,20 / Math: 80,70,75の場合、一見Math80点が高そうに見えるが、平均を見るとEng: 30, Math:75なので、Eng60点の方が良い点と判断できる。
- 特徴量によって桁違いにサイズが異なるデータセットの場合もあり、線形モデルやカーネル法を用いたSVMでも正しい学習ができない。
- 正規分布が前提となる回帰分析などでは、対数正規分布にしてから学習したほうが良い。

# Feature Extraction

**【課題】PCAの実行**

```python
data_fs.head()
```

```python
from sklearn.decomposition import PCA

#data_fsに対しPCAを行い分散説明率の累積を確認

pca = PCA(n_components=None)
X_train_pca = pca.fit_transform(data_fs)
var_exp = pca.explained_variance_ratio_
cum_var_exp = np.cumsum(var_exp)

import matplotlib.pyplot as plt
plt.bar(range(len(data_fs.columns)), var_exp, alpha=0.5, align='center', label='individual explained variance')
plt.step(range(len(data_fs.columns)), cum_var_exp, where='mid', label='cumulative expalined variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.show()
```

```python
#分散説明率が70%を超える数の主成分を選び、data_pcaに格納する

pca = PCA(n_components=2) 
data_pca = pca.fit_transform(data_fs)
```

**【課題】なぜPCAを行うのか記述せよ**

メリットは2点。
 - 計算量の削減

    特徴量は少なければ少ない方に越したことはない。無駄な処理を防ぐため特徴量選択が必要。


- 過学習（次元の呪い）の回避

    次元が増えると空間上の点の組み合わせが増えていくことが、次元の呪い。対応方法は、膨大なデータを準備するか、特徴量を減らすか、になるが、データ取得には限界があるため、特徴量を減らすことが有効とされている。

# クラスタリング

**【課題】クラスタリングとは何か記述せよ**

- いつ何に使うのか

    教師なし学習の中でも、データを自然なまとまりに分けたい場合。

- 階層的手法と非階層的手法との違い

    階層的手法は、樹形図として意味のある分割を作成する。クラスタの個数を事前に指定する必要がない。凝集型と分岐型の2つのアプローチがある。計算量が少ないために凝集型が使われることが多い。
       - 凝集型では、ボトムアップ的に階層を作る。
       - 分岐型では、トップダウン的に分割して、階層を作る。

    非階層的手法は、評価関数（ex. クラスタ内誤差平方和、SSE）を定義し、最適になるような分割を行う手法。エルボー法やシルエット図を使い、クラスタリングの品質検証する。代表的な手法は、K-means法。


- クラス分類（classification）との比較

    クラス分類は教師あり学習、あらかじめ分類するグループが決まっている。クラスタリングは教師なし学習で、グループの分け方はその時の学習次第。
    クラスタリングでは、人間が与えるラベルに近いかどうかではなく、データと各アルゴリズムに基づいてグループ分けや回帰を行う。

**【課題】k-meansにおけるkの値の検討**

```python
#エルボー図
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

SSE = []
for i in range(2, 20):
    km = KMeans(n_clusters=i, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0)
    km.fit(data_pca)
    SSE.append(km.inertia_)

plt.plot(range(2,20), SSE, marker='o')
plt.xlabel('k')
plt.ylabel('SSE')#残差平方和
plt.show()
```

```python
# シルエット図

from matplotlib import cm
from sklearn.metrics import silhouette_samples
import numpy as np

for index in range(15,18):
    km = KMeans(n_clusters=index, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0)
    y_km = km.fit_predict(data_pca)

    cluster_labels = np.unique(y_km)
    n_clusters = cluster_labels.shape[0]
    silhouette_vals = silhouette_samples(data_pca, y_km, metric='euclidean')
    y_ax_lower, y_ax_upper = 0, 0
    yticks = []

    for i, c in enumerate(cluster_labels):
        c_silhouette_vals = silhouette_vals[y_km == c]
        c_silhouette_vals.sort()
        y_ax_upper += len(c_silhouette_vals)
        color = cm.jet(i / n_clusters)
        plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, edgecolor='none', color=color)
        yticks.append((y_ax_lower + y_ax_upper) / 2)
        y_ax_lower += len(c_silhouette_vals)

    silhouette_avg = np.mean(silhouette_vals)
    plt.axvline(silhouette_avg, color="red", linestyle="--")
    plt.yticks(yticks, cluster_labels + 1)
    plt.ylabel('Cluster')
    plt.xlabel('Silhouette coefficient '+str(index))
    plt.show()
```

クラスター数

k=15<br>
理由：エルボー図からk=15から残差平方和が収束しているため。シルエット図からもサンプルの凝集度がk=16よりもそれぞれのクラスタで形が均等と判断し、k=15を採用。

**【課題】k-meansの実行**

```python
from sklearn.cluster import KMeans

k = 15
km = KMeans(n_clusters=k, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0)
y_km = km.fit_predict(data_pca)
```

**【課題】クラスタリングの可視化**

```python
#散布図を描くコードを記述
#中心点もプロットする

plt.figure(figsize =(15,10))
plt.title('y_km')
i = 1
plt.subplot(1,1,1)
for y_km_i in range(k):

    c = cm.spring(float(i) / k)
    plt.scatter(data_pca[y_km==y_km_i,0], 
                data_pca[y_km==y_km_i,1], 
                c=c , 
                marker='o', 
                s=50, 
                label='k='+str(y_km_i+1))

    plt.scatter(km.cluster_centers_[:,0],
                km.cluster_centers_[:,1],
                c=c , 
                marker='*', 
                s=250, 
                label='centroids for k='+str(y_km_i+1))
    i+=1

plt.legend()
plt.grid()
plt.show()
```
