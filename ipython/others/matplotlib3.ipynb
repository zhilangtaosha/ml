## ２次元ヒストグラム

```python
mean = [0,0]
cov = [[1,1],[1,2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

## 等高線

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
print(x.shape)
print(y.shape)
print(X.shape)
print(Y.shape)
```

```python
Z = np.sin(X)** 10 + np.cos(10 + Y +X) * np.cos(X)
plt.contour(X, Y, Z,color='black')
```

```python
plt.contour(X, Y, Z, 30, cmap='RdGy');
```

```python
import numpy as np
import math

# f(x)を自分で書く
f = lambda x: (math.exp(-x**2/2)) / math.sqrt(2*math.pi)

# ベクトルxを [-5.0, ..., 5.0] の区間で作成
n = np.linspace(-5.0, 5.0, 10000)

# f(x)の結果を得る
p = []
for i in range(len(n)):
    p.append(f(n[i]))

# グラフに表示
plt.scatter(n, p)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
mu, sd = 20, 5

s = np.random.normal(mu, sd, 10000)

count, bins, ignored = plt.hist(s, 40, normed=True)
print(count)
print(bins)

plt.plot(bins, 1/(sd * np.sqrt(2 * np.pi)) *np.exp( - (bins - mu)**2 / (2 * sd**2) ),linewidth=2, color='r')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math

mu = 0                      # 平均
variance = 0.7              # 分散
sigma = math.sqrt(variance) # 標準偏差(分散の平方根)

x = np.linspace(mu-3*variance,mu+3*variance, 100)

plt.plot(x,mlab.normpdf(x, mu, sigma)) # normpdf:確率密度関数

plt.show()
```
