```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.plot(x, y ,'o', color='blue')
```

```python
r = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(r.rand(5), r.rand(5),  marker, label='marker="{0}"'.format(marker))

plt.legend(numpoints = 1)
plt.xlim(0, 2);
```

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.plot(x, y , '-o', color='blue')
```

```python
plt.plot(x, y, '-p', color='pink',
         markersize=10, linewidth=20,
         markerfacecolor='red',
         markeredgecolor='green',
         markeredgewidth=3
        )
plt.ylim(-1.2, 1.2);
```

# scatter

```python
plt.scatter(x, y, marker='o');
```

```python
rnd = np.random.RandomState(0)

x = rnd.randn(1000)
y = rnd.randn(1000)
colors = rnd.rand(1000)
sizes = 1000* rnd.rand(1000)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.2,
            cmap='viridis')
plt.colorbar();
```
