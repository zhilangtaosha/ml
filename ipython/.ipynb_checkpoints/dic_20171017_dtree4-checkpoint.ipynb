```python
import numpy as np
from sklearn.datasets import load_boston
boston = load_boston()
import pandas as pd
X=pd.DataFrame(boston.data[:100,:], columns=boston.feature_names)
y=boston.target[:100]
```

```python
x = X['LSTAT'].values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.scatter(x,y)
```

# モデル構築
次にモデルの構築です。モデルの構築は、scikit-learnのDecisionTreeRegressorクラスを用いて行うことが可能です。木の分岐の深さは3にします。


```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
from io import StringIO
from operator import itemgetter


tree_3d = DecisionTreeRegressor(max_depth = 3)
tree_3d.fit(x[:,None], y)
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

def main():
    dataset = datasets.load_iris()

    features = dataset.data
    targets = dataset.target

    # Petal length と Petal width だけを特徴量として使う
    petal_features = features[:, 2:]

    # モデルを学習させる
    clf = DecisionTreeClassifier(max_depth=3)
    clf.fit(petal_features, targets)

    # DOT 言語のフォーマットで決定木の形を出力する
    with open('iris-dtree.dot', mode='w') as f:
        tree.export_graphviz(clf, out_file=f)


if __name__ == '__main__':
    main()
```

```python
# 回帰直線を図示するのに変数を並び替え
sort_idx=x[:,None].flatten().argsort()

# プロット
plt.figure(figsize = (10, 7))

plt.plot(x[sort_idx], tree_3d.predict(x[sort_idx].reshape(100,1)), color = 'red', label = 'depth = 3')

plt.scatter(x[sort_idx], y[sort_idx], c = 'blue', label = 'Training points')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.legend(loc = 'upper right')
plt.show()
```

今回のテキストでは、LSTATのデータ100個に対し、木の深さを3に設定しました。テキストで学んだscikit-learnで使った手法で、木の深さを1,3,5,7,10と変えてプロットさせてみましょう。

```python
dep = [1,3,5,7,10]
y_plot = []

for di in dep:
    tree_ = DecisionTreeRegressor(max_depth = di)
    tree_.fit(x[:,None], y)

    # 回帰直線を図示するのに変数を並び替え
    sort_idx=x[:,None].flatten().argsort()

    y_ = tree_.predict(x[sort_idx].reshape(100,1))
    y_plot.append(y_)
```

```python
from matplotlib import cm

plt.figure(figsize = (20, 7))
#total_label = len(y_plot)
j=1
plt.subplot(1,2,1)
for i, y__ in zip(dep[:3], y_plot[:3]):
    # プロット
    c = cm.spring(float(j) / 4)
    plt.plot(x[sort_idx], y__, color = c, label= str(i) + 'depth_dtree')
    j += 1

plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.legend(loc = 'upper right')
plt.scatter(x[sort_idx], y[sort_idx], c = 'blue', label = 'Training points')

j=1
plt.subplot(1,2,2)
for i, y__ in zip(dep[3:], y_plot[3:]):
    # プロット
    c = cm.spring(float(j) / 2)
    plt.plot(x[sort_idx], y__, color = c, label= str(i) + 'depth_dtree')
    j += 1

plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.legend(loc = 'upper right')
plt.scatter(x[sort_idx], y[sort_idx], c = 'blue', label = 'Training points')
plt.show()
```

```python
# bootstrap sample
#袋からボールを取り出すプログラム
import random

colors = ['赤','ピンク','緑','青','黄']
balls_bag = [0,1,2,3,4] #文字列はrandomモジュールで扱えないので番号をつけておく
A = []
B = []
C = []
D = []

def ball_choice(balls_bag,people,n):
    tmp = random.sample(balls_bag, n) #nコ袋から取る
    for i in tmp:
        #番号に対応する色を記録する
        people.append(colors[i])

ball_choice(balls_bag,A,3)
ball_choice(balls_bag,B,3)
ball_choice(balls_bag,C,3)
ball_choice(balls_bag,D,3)

print("Aさん",A," Bさん",B," Cさん",C," Dさん",D)
```
