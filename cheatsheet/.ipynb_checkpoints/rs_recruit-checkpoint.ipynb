# Preprocessing
* * * * *

## The total number of reservation

#### Count total num of reserve of air

```python
a_reserve_update = a_reserve.set_index(a_reserve.visit_datetime)
a_reserve_update.index = pd.to_datetime(a_reserve_update.index, utc=True)
a_reserve_grpby = a_reserve_update.groupby(["air_store_id"]).resample('D')['reserve_visitors'].sum()
print(a_reserve_grpby.head())
print("_"*30)
print(a_reserve_grpby.count())
```

#### Count total num of reserve of hpg

```python
h_reserve_update = h_reserve.set_index(h_reserve.visit_datetime)
h_reserve_update.index = pd.to_datetime(h_reserve_update.index, utc=True)
#h_reserve_update.index = pd.to_datetime(h_reserve_update.index)
h_reserve_grpby = h_reserve_update.groupby(["hpg_store_id"]).resample('D')['reserve_visitors'].sum()
print(h_reserve_grpby.head())
print("_"*30)
print(h_reserve_grpby.count())
```

#### Merge

```python
# Prepare data to merge with air_id as key
a_reserve_grouped = pd.DataFrame(a_reserve_grpby).reset_index()
h_reserve_grouped = pd.DataFrame(h_reserve_grpby).reset_index()

#a_reserve_grouped.to_csv("a_reserve_grouped.csv")
#h_reserve_grouped.to_csv("h_reserve_grouped.csv")
```

```python
import csv
with open("../../../mltestdata/05_recruit/store_id_relation.csv") as f:
    store_mapping = dict(filter(None, csv.reader(f)))

store_mapping_inv = {v:k for k, v in store_mapping.items()}

h_reserve_grouped['air_store_id'] = h_reserve_grouped['hpg_store_id'].map(store_mapping_inv)
```

```python
h_reserve_grouped.head()
```

```python
h_reserve_grouped = h_reserve_grouped.dropna(subset=['air_store_id'])
```

```python
h_reserve_grouped.head()
```

```python
# Merge
total_reserve = a_reserve_grouped.rename(columns={'reserve_visitors': 'a_res_vis'})
print(total_reserve.count())
total_reserve.to_csv("total_reserve.csv")
```

```python
total_reserve.head()
```

```python
h_reserve_grouped.head()
```

```python
total_merged = pd.merge(total_reserve, h_reserve_grouped, on=['air_store_id', 'visit_datetime'], how='outer')
#total_merged = pd.merge(total_reserve, h_reserve_grouped, on=['air_store_id', 'visit_datetime'])
```

```python
total_merged.head()
```

```python
total_merged = total_merged.drop(['hpg_store_id'], axis=1)
```

```python
total_merged = total_merged.rename(columns={'reserve_visitors' : 'h_res_vis'})
print(total_merged.count())
```

```python
total_merged.head()
```

```python
total_merged = total_merged.fillna(0)
total_merged['total_res_vis'] = total_merged.a_res_vis + total_merged.h_res_vis
#total_merged.sort_values(by=["air_store_id","visit_datetime"], ascending=True)
```

```python
total_merged.head()
```

```python
total_merged.visit_datetime = total_merged.visit_datetime.apply(lambda x: x.strftime('%Y-%m-%d'))
total_merged = total_merged.rename(columns={'visit_datetime' : 'visit_date'})
```
